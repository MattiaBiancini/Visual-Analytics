### Step 1

After filtering all the data, I am going to show the city with the most accidents both years.

There are several ways to decide how many cities should I plot. Some of them are:
- 80/20 Rule (or Pareto Principle)
- Cumulative Sum
- Percentage number (Top N% Cities)
- Fixed number (Top N Cities)

```python
df_accidents['Year'] = df_accidents['Start_Time'].dt.year
accidents_per_year = df_accidents.groupby(['Year', 'City']).size().sort_values(ascending=False)
accidents_per_year
```

#### 80/20 Rule
With this rules I should take the 20% of the city.

```python
accidents_8020 = {y: accidents_per_year.loc[y][:int(len(accidents_per_year.loc[y]) * 0.2)] for y in years}
for y in years:
    print(f"Number of cities using 80/20 rule in the year {y}: {len(accidents_8020[y])}")
```

Representing about `3000` cities is not possible since the number is way to big even for a table.

#### Cumulative Sum
The cumulative sum allows to see when the increment is not that meaningful and I can extract a good percentage of the whole dataset (in this case `80%` of the total number of accidents).

```python
for year in years:
    # Filter for the chosen year and sort descending
    accidents_year = accidents_per_year.loc[year].sort_values(ascending=False)

    # Calculate cumulative sum
    cumulative_accidents = accidents_year.cumsum()

    # Total number of accidents for the year
    total_accidents = cumulative_accidents.max()

    # Calculate percentage of total accidents
    percentage = cumulative_accidents / total_accidents * 100

    # Set the threshold (e.g., 80% of total accidents)
    threshold = 80  # Change to 90 or other as needed

    # Find where the cumulative sum reaches the threshold
    cut_off = np.argmax(percentage >= threshold) + 1  # +1 because of zero-indexing

    # Plot the cumulative sum curve
    plt.figure(figsize=(10, 6))
    plt.plot(np.arange(1, len(cumulative_accidents) + 1), 
             cumulative_accidents, 
             marker='o', 
             markersize=2,  # Thinner dots
             label='Cumulative Accidents')

    # Vertical line for threshold cut-off
    plt.axvline(x=cut_off, color='r', linestyle='--', 
                label=f'Cut-off at {cut_off} cities ({threshold}%)')

    # Title and axis labels
    plt.title(f'Cumulative Sum of Accidents by Number of Cities ({year})')
    plt.xlabel('Number of Cities Considered')
    plt.ylabel('Number of Accidents')

    # Format y-axis with commas (e.g., 000,000)
    plt.gca().yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f'{int(x):,}'))

    # Grid and legend
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Print cut-off information
    print(f"Cut-off at {cut_off} cities covers {threshold}% of total accidents in {year}.")
```

Using the Cumulative Sum the number of city to represent is decreased to `2000`, but this is still a huge number to be visualized.

Now I am going to show the top `10` cities in both 2020 and 2022

```python
def create_hplot(year, top=10, max_range=0, color="steelblue"):
    # Get top 10 accidents for the year
    accidents_top10 = accidents_per_year.loc[year][0:top].sort_values(ascending=True)
    
    # Plot title and axis labels
    title = f"Top 10 US Cities by Number of Accidents in {year}"
    x_axis_label = "Number of Accidents"
    y_axis_label = "US Cities"

    # Data source for Bokeh
    source = ColumnDataSource(data={
        'city': accidents_top10.index.astype(str),      # City names
        'accidents': accidents_top10.values            # Accident numbers
    })

    # Tooltip format
    TOOLTIPS = f"""
    <div style="background-color: #f9f9f9; padding: 8px; border-radius: 8px;">
        <span style="font-size: 12px; color: {colors[year]};">City:</span> @city<br>
        <span style="font-size: 12px; color: {colors[year]};">Accidents:</span> @accidents{0,0}<br>
    </div>
    """

    # Define y-axis range
    y_range_list = accidents_top10.index.astype(str).tolist()

    # Create Bokeh plot
    plot = figure(
        y_range=y_range_list,
        height=400,
        width=700,
        tooltips=TOOLTIPS,
        tools="save",
        title=title,
        x_axis_label=x_axis_label,
        y_axis_label=y_axis_label
    )

    # Horizontal bar plot with dynamic color
    plot.hbar(y='city', right='accidents', source=source, height=0.85, color=color)
    
    # Grid and axis formatting
    plot.xgrid.grid_line_color = None
    plot.xaxis[0].formatter = NumeralTickFormatter(format="0,0")  # Format large numbers
    plot.yaxis.minor_tick_line_color = None
    plot.xaxis.minor_tick_line_color = None
    plot.x_range.start = 0
    plot.x_range.end = max_range  # Same range for all plots

    return plot
```

```python
def round_to_nearest(value, base=5000):
    return int(base * math.ceil(value / base))

max_range = round_to_nearest(int(accidents_per_year.max()))
plots = [create_hplot(year, max_range=max_range, color=colors[year]) for year in years]

layout = column(*plots)

show(layout)
```

In this case I consider that if an accident has a `Start_Time` in `June` but its `End_Time` is in `July` it will be considered as an accident of `June` since the instant in which it happen is in `June`.

```python
df_accidents['Month'] = df_accidents['Start_Time'].dt.month
df_accidents['Duration'] = (df_accidents['End_Time'] - df_accidents['Start_Time']).dt.total_seconds() / 60
df_accidents['Duration'] = df_accidents.Duration.apply(lambda x : int(x))

df_accidents_sccym = df_accidents.groupby(['State', 'County', 'City', 'Year', 'Month'])['Duration'].agg(['min', 'max']).reset_index()
df_accidents_sccym.rename(columns={'min' : 'Min', 'max': 'Max'}, inplace=True)

valid_entries = df_accidents_sccym.groupby(['State', 'County', 'City', 'Year'])['Month'].nunique() == 12
df_accidents_sccym = df_accidents_sccym.set_index(['State', 'County', 'City', 'Year']).loc[valid_entries].reset_index()

df_accidents_sccym
```

```python
from bokeh.models import ColumnDataSource, Span, NumeralTickFormatter, HoverTool
from bokeh.plotting import figure, show
from bokeh.layouts import row, column
import calendar

# Calculate means
cities = ['Los Angeles', 'Dallas', 'New York']
mins = {city: float(df_accidents_cities[df_accidents_cities['City'] == city]['Min'].mean()) for city in cities}
maxs = {city: float(df_accidents_cities[df_accidents_cities['City'] == city]['Max'].mean()) for city in cities}

# Global min/max for y_range
global_min_y = (df_accidents_cities['Min'].min(), int(df_accidents_cities['Min'].max()*1.1))
global_max_y = (df_accidents_cities['Max'].min(), int(df_accidents_cities['Max'].max()*1.1))

# Create plot function
def create_city_plot(df, city, year, duration_type='Min'):
    """
    Creates a Bokeh plot for accident durations by month with consistent y-axis.
    
    Parameters:
    - df (DataFrame): The input DataFrame with accident data.
    - city (str): The city to filter.
    - year (int): The year to filter.
    - duration_type (str): 'Min' or 'Max' for duration type.
    """
    # Filter data
    city_data = df[(df['City'] == city) & (df['Year'] == year)].copy()
    city_data['Month Name'] = city_data['Month'].apply(lambda x: calendar.month_abbr[x])
    
    source = ColumnDataSource(city_data)
    
    duration_label = 'minimum' if duration_type == 'Min' else 'maximum'
    title = f"{city} {duration_label} accident duration in {year}"
    
    x_range = sorted(city_data['Month Name'].unique(), key=lambda m: list(calendar.month_abbr).index(m))
    
    # Y-axis range based on global min/max values
    y_range = global_min_y if duration_type == 'Min' else global_max_y
    
    # Get mean value based on duration type
    mean_value = mins[city] if duration_type == 'Min' else maxs[city]
    
    # Create Bokeh plot
    p = figure(
        title=title,
        x_range=x_range,
        y_range=y_range,
        x_axis_label=None,
        y_axis_label="Duration (minutes)",
        width=500,
        height=300
    )
    
    # Add HoverTool with custom duration formatting
    hover = HoverTool(tooltips=[("Month", "@{Month Name}"),
                                ("Duration", "@{" + duration_type + "} min")])
    p.add_tools(hover)
    
    # Line and scatter plot
    p.line(x='Month Name', y=duration_type, source=source, line_width=2, color='steelblue')
    p.scatter(x='Month Name', y=duration_type, source=source, size=5, alpha=1, color='steelblue')
    
    # Add horizontal mean line
    mean_line = Span(location=mean_value, dimension='width', 
                     line_color='orange', line_dash='dashed', line_width=2)
    p.add_layout(mean_line)
    
    # Style adjustments
    p.toolbar.logo = None
    p.toolbar_location = None
    p.ygrid.grid_line_color = None
    p.yaxis[0].formatter = NumeralTickFormatter(format="0")
    p.yaxis.minor_tick_line_color = None
    p.xaxis.minor_tick_line_color = None
    p.y_range.start = 0
    
    return p

# Create plots with shared y-axis ranges
columns = []

for city in cities:
    columns.append(column(
        create_city_plot(df_accidents_cities, city, 2022, 'Min'),
        create_city_plot(df_accidents_cities, city, 2022, 'Max')
    ))

# Show the grid layout
show(row(*columns))
```
